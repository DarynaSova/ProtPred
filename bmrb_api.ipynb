```python
import requests
import pandas as pd
from io import StringIO
response = requests.get("http://api.bmrb.io/current/mappings/bmrb/pdb")
```

```python

data = response.json()
print(data[:500]) 
```

```python
rows = []
for entry in data:
    bmrb_id = entry["bmrb_id"]
    for pdb_id in entry["pdb_ids"]:
        rows.append({"bmrb_id": bmrb_id, "pdb_id": pdb_id})


df = pd.DataFrame(rows)
```
